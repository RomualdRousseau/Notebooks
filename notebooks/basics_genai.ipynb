```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
embeds = {
    "king": [1., 0],
    "queen": [1., 0.5],
    "bed": [0., 1.]
}

x, y = list(zip(*embeds.values(), strict=False))
plt.scatter(x, y)
for x_, y_, k_ in zip(x, y, embeds.keys(), strict=False):
    plt.text(x_, y_, k_)
plt.show()
```

```python
def softmax(v):
    exp_vector = np.exp(v)
    return exp_vector / np.sum(exp_vector)

input = np.array([[0], [0], [1]])
output = np.array([[0], [1], [1]])

E = np.array(list(embeds.values()))
d = np.sqrt(E.shape[1])
Q = input * E
K = (input * E).T
V = output * E
H = V @ softmax(((K @ Q) / d))

x, y = list(zip(*H, strict=False))
plt.scatter(x, y)
for x_, y_, k_ in zip(x, y, embeds.keys(), strict=False):
    plt.text(x_, y_, k_)
plt.show()
```
