```python
import os

import kagglehub
from groq import Groq
from IPython.display import Markdown
```

```python
df = kagglehub.dataset_load(
  kagglehub.KaggleDatasetAdapter.PANDAS,
  "shayanhusain/drug-food-interactions-dataset",
  "Drug to Food interactions Dataset.json",
)

df.head()
```

```python
client = Groq(
    api_key=os.environ['GROQ_API_KEY'],
)
```

```python
chat_completion = client.chat.completions.create(
    messages=[
        {"role": "user", "content":  f"""
        You are a medical expert but not a doctor.
        You answer the question about food to avoid when under medical treatment. Try to find example a disease linked to the drug.

        Data available:
        {df}

        Question:

        I love drink beer. What disease should I avoid?
        """}
    ],
    model="llama-3.3-70b-versatile",
)

narrative = chat_completion.choices[0].message.content

Markdown(narrative)
```
